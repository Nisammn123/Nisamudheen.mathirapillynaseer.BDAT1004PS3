```python
#Question 1
# Step 1: Import the necessary libraries
import pandas as pd
import numpy as np

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

# Step 3: Assign it to a variable
users = pd.read_csv(url, sep='|') 

# Step 4: Mean age per occupation
print("Mean Age per Occupation:\n")
mean = users.groupby('occupation')['age'].mean()
print(mean)
print("\n")  

# Step 5: Male ratio per occupation
print("Male Ratio per Occupation (sorted from most to least):\n")
users['male_ratio'] = users['gender'].apply(lambda x: 1 if x == 'M' else 0)
ratio = users.groupby('occupation')['male_ratio'].mean().sort_values(ascending=False)
print(ratio)
print("\n")

# Step 6: Minimum and maximum ages per occupation
print("Minimum and Maximum Ages per Occupation:\n")
min_max = users.groupby('occupation')['age'].agg(['min', 'max'])
print(min_max)
print("\n")

# Step 7: Mean age by occupation and gender
print("Mean Age by Occupation and Gender:\n")
mean_age = users.groupby(['occupation', 'gender'])['age'].mean()
print(mean_age)
print("\n")

# Step 8: Percentage of men and women per occupation
print("Percentage of Men and Women per Occupation:\n")
gender_count = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)
gender_percentage = gender_count.div(gender_count.sum(axis=1), axis=0) * 100
print(gender_percentage)
print("\n")

```

```python
# Question 2
# Step 1: Import the necessary libraries
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"

# Step 3: Assign it to a variable
euro12 = pd.read_csv(url)

# Step 4: Select only the Goal column
print("Goals Column:\n")
goals = euro12['Goals']
print(goals)
print("\n")

# Step 5: How many teams participated in Euro 2012?
print("Number of Teams Participated in euro12:\n")
number = euro12['Team'].nunique()
print(number)
print("\n")

# Step 6: What is the number of columns in the dataset?
print("Number of Columns in the Dataset:\n")
num_columns = euro12.shape[1]
print(num_columns)
print("\n")

# Step 7: View columns Team, Yellow Cards, and Red Cards
print("Team, Yellow Cards, and Red Cards:\n")
view = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print(view)
print("\n")

# Step 8: Sort teams by Red Cards, then Yellow Cards
print("Teams Sorted by Red Cards and Yellow Cards:\n")
sorted_teams = view.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=[False, False])
print(sorted_teams)
print("\n")

# Step 9: Calculate the mean Yellow Cards per team
print("Mean Yellow Cards per Team:\n")
mean = euro12['Yellow Cards'].mean()
print(mean)
print("\n")

# Step 10: Filter teams that scored more than 6 goals
print("Teams that Scored More Than 6 Goals:\n")
teams_more_than_6 = euro12[euro12['Goals'] > 6]['Team']
print(teams_more_than_6)
print("\n")

# Step 11: Select teams that start with the letter 'G'
print("Teams that Start with 'G':\n")
teams_starting_g = euro12[euro12['Team'].str.startswith('G')]['Team']
print(teams_starting_g)
print("\n")

# Step 12: Select the first 7 columns
print("First 7 Columns:\n")
first_7_columns = euro12.iloc[:, :7]
print(first_7_columns)
print("\n")

# Step 13: Select all columns except the last 3
print("All Columns Except the Last 3:\n")
except_last_3 = euro12.iloc[:, :-3]
print(except_last_3)
print("\n")

# Step 14: Present Shooting Accuracy for England, Italy, and Russia
print("Shooting Accuracy for England, Italy, and Russia:\n")
shooting_accuracy = euro12.loc[euro12['Team'].isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']]
print(shooting_accuracy)
print("\n")
```

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np

# Step 2: Make 3 series with 100 random numbers
beds = pd.Series(np.random.randint(1, 5, size=100))  # Random numbers from 1 to 4
baths = pd.Series(np.random.randint(1, 4, size=100))  # Random numbers from 1 to 3
prices = pd.Series(np.random.randint(10000, 30001, size=100))  # Random numbers from 10,000 to 30,000

# Step 3: Join series into a DataFrame
data = pd.concat([beds, baths, prices], axis=1)

# Step 4: Rename columns
data.columns = ['beds', 'baths', 'price_per_sqm']
print("Data with renamed columns:\n")
print(data.head())  # Show first 5 rows
print("\n")

# Step 5: Combine all series into one column
big_data = pd.concat([beds, baths, prices], ignore_index=True)
big_data = pd.DataFrame(big_data, columns=['big_column'])
print("Single column of combined data:\n")
print(big_data.head(110))  # Show first 110 rows
print("\n")

# Step 6: Check if index stops at 99
print("Does it stop at 99?\n")
print(big_data.index.max())  # Show max index
print("\n")

# Step 7: Reindex to go from 0 to 299
big_data = big_data.reindex(range(300))
print("Reindexed data:\n")
print(big_data.tail(10))  # Show last 10 rows
print("\n")
```

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np


# Step 2: Load the data
# I'm loading the wind data from the provided file
data = pd.read_csv("wind.txt", sep=r'\s+', skiprows=1, header=None)

# Assigning column names
data.columns = ['year', 'month', 'day', 'RPT', 'VAL', 'ROS', 'KIL', 'SHA', 'BIR',
                'DUB', 'CLA', 'MUL', 'CLO', 'BEL', 'MAL']

# Step 3: Replace the first three columns with a datetime index
# I'll name the columns and combine the year, month, and day into a proper date
data[['year', 'month', 'day']] = data[['year', 'month', 'day']].astype(int)

# Combine year, month, and day into a proper date
data['date'] = pd.to_datetime(data[['year', 'month', 'day']].assign(year=lambda x: x['year'] + 1900))

# Drop original year, month, and day columns and set date as index
data = data.drop(['year', 'month', 'day'], axis=1)
data = data.set_index('date')

# Step 4: Fix the year 2061 issue
# I'll write a simple function to fix years and apply it
def fix_years(date):
    if date.year > 2000:
        return date.replace(year=date.year - 100)
    return date

# Applied the function to fix the index
data.index = data.index.map(fix_years)

# Step 5: Check the index type
# Making sure the index is datetime
data.index = pd.to_datetime(data.index)

# Step 6: Count missing values for each location
print("1.Missing values per location:\n")
missing_values = data.isnull().sum()
print(missing_values)
print("\n")

# Step 7: Count total non-missing values
print("2.Total non-missing values:\n")
non_missing = data.notnull().sum().sum()
print(non_missing)
print("\n")

# Step 8: Mean windspeed for the whole dataset
print("3.Overall mean windspeed:\n")
mean_windspeed = data.mean().mean()
print(mean_windspeed)
print("\n")

# Step 9: Stats (min, max, mean, std) for each location
print("4.Stats per location:\n")
loc_stats = data.describe().T[['min', 'max', 'mean', 'std']]
print(loc_stats)
print("\n")

# Step 10: Stats (min, max, mean, std) per day across locations
print("5.Stats per day:\n")
day_stats = data.aggregate(['min', 'max', 'mean', 'std'], axis=1)
print(day_stats.head())
print("\n")

# Step 11: Average windspeed in January for each location
print("6.Average windspeed in January:\n")
january_data = data[data.index.month == 1]
jan_avg = january_data.groupby(january_data.index.year).mean()
print(jan_avg)
print("\n")

# Step 12: Downsample to yearly frequency
print("7.Yearly stats for each location:\n")
yearly = data.resample('YE').mean()  # Use 'YE' for yearly end frequency
print(yearly)
print("\n")

# Step 13: Downsample to monthly frequency
print("8.Monthly stats for each location:\n")
monthly = data.resample('ME').mean()  # Use 'ME' for monthly end frequency
print(monthly)
print("\n")

# Step 14: Downsample to weekly frequency
print("9.Weekly stats for each location:\n")
weekly = data.resample('W').mean()
print(weekly)
print("\n")

# Step 15: Weekly stats for first 52 weeks starting Jan 2, 1961
print("10.Weekly stats (first 52 weeks):\n")
weekly_stats = data.resample('W-MON').mean().iloc[:52].describe().T[['min', 'max', 'mean', 'std']]
print(weekly_stats)
print("\n")
```

```python
# Step 1: Import libraries
import pandas as pd

# Step 2: Load the dataset
# I'm loading the Chipotle dataset from the given address
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

# Step 3: Assign it to a variable
chipo = pd.read_csv(url, sep='\t')

# Step 4: See the first 10 entries
print("1.First 10 entries:\n")
print(chipo.head(10))
print("\n")

# Step 5: Number of observations (rows)
print("2.Number of observations:\n")
print(chipo.shape[0])  # Number of rows
print("\n")

# Step 6: Number of columns
print("3.Number of columns:\n")
print(chipo.shape[1])  # Number of columns
print("\n")

# Step 7: Print the name of all the columns
print("4.Column names:\n")
print(chipo.columns)
print("\n")

# Step 8: How is the dataset indexed?
print("5.Dataset index:\n")
print(chipo.index)
print("\n")

# Step 9: Most ordered item
print("6.Most ordered item:\n")
most_ordered_item = chipo.groupby('item_name').sum().sort_values('quantity', ascending=False).head(1)
print(most_ordered_item)
print("\n")

# Step 10: Number of most-ordered items
print("7.Number of most-ordered items:\n")
print(most_ordered_item['quantity'].iloc[0])
print("\n")

# Step 11: Most ordered item in the choice_description column
print("8.Most ordered item in choice_description:\n")
most_ordered_choice = chipo.groupby('choice_description').sum().sort_values('quantity', ascending=False).head(1)
print(most_ordered_choice)
print("\n")

# Step 12: Total items ordered
print("9.Total items ordered:\n")
print(chipo['quantity'].sum())
print("\n")

# Step 13: Convert item price to float
print("10.Converting item price to float:\n")

# Remove the dollar sign and convert to float
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
print(chipo['item_price'].dtypes)  # Check the type
print("\n")

# Step 14: Total revenue
print("11.Total revenue:\n")
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
print(chipo['revenue'].sum())
print("\n")

# Step 15: Total orders
print("12.Total orders:\n")
print(chipo['order_id'].nunique())
print("\n")

# Step 16: Average revenue per order
print("13.Average revenue per order:\n")
average_revenue = chipo['revenue'].sum() / chipo['order_id'].nunique()
print(average_revenue)
print("\n")

# Step 17: Number of different items sold
print("14.Number of different items sold:\n")
print(chipo['item_name'].nunique())
print("\n")
```

```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset
# I'll load the dataset from the local file
file = "us-marriages-divorces-1867-2014.csv"
data = pd.read_csv(file)

# Step 3: Create the line plot
plt.figure(figsize=(10, 6))

# Plot marriages and divorces per 1000 people
plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages per 1000', linestyle='-', marker='o')
plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces per 1000', linestyle='--', marker='x')

# Step 4: Add labels, legend, and title
plt.xlabel('Year')
plt.ylabel('Rate per 1000 People')
plt.title('Marriages and Divorces per Capita in the U.S. (18672014)')
plt.legend()

# Step 5: Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Question 7

import pandas as pd
import matplotlib.pyplot as plt

# loaded the dataset from the local file
file_path = "us-marriages-divorces-1867-2014.csv"
data = pd.read_csv(file_path)

# Filter data for 1900, 1950, and 2000
# select only the years interested in
F_data = data[data['Year'].isin([1900, 1950, 2000])]

# Create the bar chart
# I'll create a bar chart to compare marriages and divorces
x_labels = F_data['Year']  # Years for the x-axis
marriages = F_data['Marriages_per_1000']  # Marriages data
divorces = F_data['Divorces_per_1000']  # Divorces data

# Set bar width
bar_width = 0.35

# Plot the bars
plt.figure(figsize=(8, 6))
x = range(len(x_labels))
plt.bar(x, marriages, width=bar_width, label='Marriages per 1000', color='blue')
plt.bar([p + bar_width for p in x], divorces, width=bar_width, label='Divorces per 1000', color='orange')

# Add labels, legend, and title
plt.xlabel('Year')
plt.ylabel('Rate per 1000 People')
plt.title('Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.xticks([p + bar_width / 2 for p in x], x_labels)  # Center the labels
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Question 8

import pandas as pd
import matplotlib.pyplot as plt

# Loaded the Roman Emperor dataset from the local file
file = "roman-emperor-reigns.csv"
data = pd.read_csv(file)

# counting how many emperors died by each cause
assassination_data = data['Cause_of_Death'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(assassination_data, labels=assassination_data.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen'])

# Adding a title and making the pie chart a circle
plt.title('Fraction of Roman Emperors Assassinated')
plt.axis('equal')  # Makes sure it looks like a circle

# Result
plt.tight_layout()
plt.show()
```
